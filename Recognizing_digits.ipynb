**Learn to Predict the Class**


```python
from sklearn import svm #loaded the data
```

```python
clf = svm.SVC(gamma = 0.001, C=100.) ##started the classifier #classifier over here and there are parameters for this
```

```python
clf.fit(digits.data[:-1], digits.target[:-1]) #showed some data to the classifier # fit meaning take the data and learn from it
```

```python
clf.predict(digits.data[-1:])
```
